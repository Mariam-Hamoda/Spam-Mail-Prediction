```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
```

```python
mail_data=pd.read_csv('/content/mail_data.csv')
```

```python
mail_data.head()
```

Data Processing

```python
mail_data.isnull().sum()
```

```python
mail_data['Category'].value_counts()
```

```python
X=mail_data['Message']
Y=mail_data['Category']
```

```python
#label data
encode=LabelEncoder()
```

```python
Y=encode.fit_transform(Y)
print(Y)
#0 ham
#1 spam
```

Split data for training and testing

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=3)
```

Feature Extraction

```python
words=TfidfVectorizer(min_df=1, stop_words='english',lowercase=True)
X_train_feature=words.fit_transform(X_train)
X_test_feature=words.transform(X_test)
print(X_train)
print(X_train_feature)
print(X_test)
print(X_test_feature)
```

```python
print(Y_train.dtype)
```

Training the Model

```python
model=LogisticRegression()
```

```python
model.fit(X_train_feature,Y_train)
```

Accuracy the training data


```python
train_predict=model.predict(X_train_feature)
train_Accuracy=accuracy_score(Y_train,train_predict)
print(train_Accuracy)
```

Accuracy the testing data


```python
test_predict=model.predict(X_test_feature)
test_Accuracy=accuracy_score(Y_test,test_predict)
print(test_Accuracy)
```

Making a Predictive System

```python
input_data = "URGENT! You have won a 1-week FREE membership in our £100,000 Prize Jackpot! Txt the word: CLAIM to No: 81010 T&C www.dbuk.net"
#convert to feature
input_data_feature=words.transform([input_data])
prediction=model.predict(input_data_feature)
print(prediction)

if prediction[0] == 0:
    print('Ham Mail')
else:
    print('Spam Mail')
```

```python
input_data = "HI ! How are u ? I'm Mariam ! I met u last night and u tell me u give me a Prize $100,000!"
input_data_feature=words.transform([input_data])
prediction=model.predict(input_data_feature)
print(prediction)

if prediction[0] == 0:
    print('Ham Mail')
else:
    print('Spam Mail')
```
